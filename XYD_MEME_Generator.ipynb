```python
import requests
from requests_oauthlib import OAuth1
import os
import logging

# Set up logging
logging.basicConfig(
    level=logging.INFO,
    format='%(asctime)s - %(levelname)s - %(message)s',
    filename='twitter_bot.log'
)

# Updated X API credentials
consumer_key = ''  # API Key
consumer_secret = ''  # API Key Secret
access_token = ''  # Access Token
access_token_secret = ''  # Access Token Secret

# Note: The bearer token (AAAAAAAAAAAAAAAAAAAAAFjizgEAAAAAnIjUdKla83Y5zCbCH%2BYsbIyJEyk%3DU4NY9THnSETczBRCDrjdH3kXKHBpzQub5w29ry21OsPR0Q1YNd)
# and client secret (AMyy16M5xtl7mfUtbiGLjx7J2ZD5888aX7w2W3QAtRVRQ-NPsN) are not needed for OAuth1

# Initialize OAuth1 authentication
auth = OAuth1(
    consumer_key,
    consumer_secret,
    access_token,
    access_token_secret
)

class TwitterBot:
    def __init__(self, image_path, caption="test"):
        self.image_path = os.path.abspath(image_path)
        self.caption = caption
        self.media_upload_url = "https://upload.x.com/1.1/media/upload.json"
        self.tweet_url = "https://api.x.com/2/tweets"
        self.constant_text = "YD is Live ON XRPL NOW!"
        self.hashtags = "#YD #YellowDuckie #MEMECOIN"

    def upload_media(self, image_path):
        """Upload image to X and return media_id"""
        try:
            if not os.path.exists(image_path):
                raise FileNotFoundError(f"Image file not found: {image_path}")

            with open(image_path, "rb") as image_file:
                files = {"media": image_file}
                response = requests.post(
                    self.media_upload_url,
                    auth=auth,
                    files=files
                )
                response.raise_for_status()
                media_data = response.json()
                media_id = media_data["media_id_string"]
                logging.info(f"Media uploaded successfully. Media ID: {media_id}")
                return media_id
        except Exception as e:
            logging.error(f"Media upload failed: {str(e)}")
            raise

    def post_tweet(self):
        """Post a tweet with text and image"""
        try:
            # Check if the image file exists
            if not os.path.isfile(self.image_path):
                logging.error(f"Image file not found: {self.image_path}")
                print(f"Error: Image file not found: {self.image_path}")
                return

            # Upload the image
            media_id = self.upload_media(self.image_path)

            tweet_text = f"{self.caption}\n{self.constant_text}\n{self.hashtags}"
            payload = {
                "text": tweet_text,
                "media": {
                    "media_ids": [media_id]
                }
            }

            response = requests.post(
                self.tweet_url,
                auth=auth,
                json=payload
            )
            response.raise_for_status()

            tweet_data = response.json()
            logging.info(f"Tweet posted successfully! Tweet ID: {tweet_data['data']['id']}")
            logging.info(f"Tweet text: {tweet_data['data']['text']}")
            print(f"Tweet posted successfully! Tweet ID: {tweet_data['data']['id']}")
            print(f"Tweet text: {tweet_data['data']['text']}")

        except Exception as e:
            logging.error(f"Tweet posting failed: {str(e)}")
            print(f"Error: Tweet posting failed - {str(e)}")

def main():
    bot = TwitterBot(
        image_path="img.png",  # Current directory
        caption=cap     # Default caption
    )
    logging.info("Twitter Bot starting single execution...")
    print("Twitter Bot executing...")
    bot.post_tweet()
    logging.info("Twitter Bot completed execution")
    print("Execution completed")

if __name__ == "__main__":
    main()
```
